```python
distortions = []
K = range(1,15)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(X)
    distortions.append(kmeanModel.inertia_)
```

```python
# Using the elbow method to find the optimal number of clusters
plt.figure(figsize=(16,8))
plt.plot(K, distortions)
plt.xlabel('No Of Clusters')
plt.ylabel('Intertia Score')
plt.title('The Elbow Method')
plt.grid()
plt.show()
```

```python
import sklearn.metrics as metrics
import sklearn.cluster as cluster

SK = range(2,15)
sil_score = []
for i in SK:
    labels=cluster.KMeans(n_clusters=i,init="k-means++",random_state=20).fit(X).labels_
    score = metrics.silhouette_score(X,labels,metric="euclidean",sample_size=230,random_state=20)
    sil_score.append(score)
    \
    print ("Silhouette score for k(clusters) = "+str(i)+" is "
           +str(metrics.silhouette_score(X,labels,metric="euclidean",sample_size=230,random_state=20)))
```
